# Columns explications

✅ temp

Normalized air temperature.

Scaled roughly between 0 and 1.

Higher number = warmer

✅ atemp — “Feels like” temperature

This is apparent temperature.

It tries to capture:

- wind chill
- humidity effects
- human perception of warmth

✅ casual

Number of rentals by non-registered users

✅ registered

Rentals by subscribed users

✅ cnt

Total rentals

- Mist

3️⃣ = Light bad weather
- Light snow
- Light rain
- Thunderstorm possible

- Heavy rain

# Weekends analysis

```python
weekends = df[(df['weekday'] == 6) | (df['weekday'] == 0)]
```

```python
weekends_comparison = weekends.groupby('hr')[['registered','casual']].mean()

weekends_comparison.round(1)
```

 - weekend riding shifts toward midday / afternoon
 - weekend registered usage appears to ramp up progressively toward midday rather than forming abrupt peaks
 - weekend riding is opportunity-driven, not clock-driven
 - registered riders behave like leisure riders

# Weather check

Reminder:

weathersit column:
1 = clear
2 = mist/cloudy
3 = light rain/snow
4 = heavy (rare)

```python
nice_weather = weekends[weekends['weathersit'] == 1]

nice_weather_hours = nice_weather.groupby('hr')[['registered','casual']].mean()

nice_weather_hours.round(1)
```

- registered peak ≈ hr 12
- casual peak     ≈ hr 14 -> even in perfect weather, casual riders peak slightly later
- weather influences willingness, not schedule

```python
medium_weather = weekends[weekends['weathersit'] == 2]

medium_weather_hours = medium_weather.groupby('hr')[['registered','casual']].mean()

medium_weather_hours.round(1)
```

- the shape of the day stays the same
- weather changes intensity, not timing

```python
bad_weather = weekends[weekends['weathersit'] >= 3]

bad_weather_hours = bad_weather.groupby('hr')[['registered','casual']].mean()

bad_weather_hours.round(1)
```

- demand shrinks as weather worsens
- activity waits, then concentrates into a narrower window 12 PM

# Weekdays

```python
week = df[(df['weekday'] != 6) & (df['weekday'] != 0)]
```

```python
week_comparison = week.groupby('hr')[['registered','casual']].mean()

week_comparison.round(1)
```

- two very sharp peaks, bimodal distribution aligned with workday transitions
- casual riders do NOT follow the work clock
- schedule-driven commuting behavior


```python
nice_weather = week[week['weathersit'] == 1]

nice_weather_hours = nice_weather.groupby('hr')[['registered','casual']].mean()

nice_weather_hours.round(1)
```

- the timing of the peaks did NOT move
- weather scales intensity but does not shift commuter schedules
- casual usage is governed more by free time than by weather alone

```python
medium_weather = week[week['weathersit'] == 2]

medium_weather_hours = medium_weather.groupby('hr')[['registered','casual']].mean()

medium_weather_hours.round(1)
```

- weather does NOT change commuter timing

```python
bad_weather = week[week['weathersit'] >= 3]

bad_weather_hours = bad_weather.groupby('hr')[['registered','casual']].mean()

bad_weather_hours.round(1)
```

- despite bad weather, the peaks are STILL at commuter hours
- casual riders are environment-sensitive, the demand drops significantly

# Skew

```python
skew = df['cnt'].skew().round(2)

skew
```

- positive skew (> 0)
- most hours have moderate demand
- a smaller number of hours have very high spikes

```python
checks = df.groupby('hr').agg(
    mean = ('cnt', 'mean'),
    median = ('cnt', 'median'),
    std=('cnt', 'std')
).round(2).sort_values(by='median', ascending=False)

checks['mean_median_diff'] = checks['mean'] - checks['median']

checks
```

What does mean < median suggest?

- occasional lower-than-usual values pulling the mean down, even strong commuter hours sometimes collapse
- communter peaks are high but fragile
- peak hours are the most volatile parts of the day, given their high standard deviation

# Conclusions:

⏰ Time Structure

Working days → bimodal commuter peaks

Weekends → broad midday plateau

🌦️ Environmental Effect

Weather scales how many people ride

But barely changes when they ride

👥 Rider Segmentation

Registered → constraint-driven

Casual → environment-driven


2 systems of riding:

🧠 System 1 — Constraint-Driven (Registered Riders)


Sharp peaks at 7–9 and 16–18

Peaks persist even under bad weather

Mean < median in commuter hours → occasional disruption, but strong baseline


👉 These riders follow external schedules.

The system behaves like a transport network during working days.



🌿 System 2 — Opportunity-Driven (Casual Riders)


Broad midday hill

Strong weather sensitivity

Larger variability and elastic hours


👉 These riders respond to comfort, leisure time, and environment.

The same bikes behave like a recreational system at other times.
